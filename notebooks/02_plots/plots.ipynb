```python
import matplotlib.pyplot as plt
import numpy as np

from flavio.statistics.functions import delta_chi2
import scipy
plt.rcParams['text.usetex'] = True
```

```python
colors_line = [(0.8941176470588236, 0.10196078431372549, 0.10980392156862745),
 (0.21568627450980393, 0.49411764705882355, 0.7215686274509804),
 (0.30196078431372547, 0.6862745098039216, 0.2901960784313726),
 (0.596078431372549, 0.3058823529411765, 0.6392156862745098),
 (1.0, 0.4980392156862745, 0.0),
 (1.0, 1.0, 0.2),
 (0.6509803921568628, 0.33725490196078434, 0.1568627450980392),
 (0.9686274509803922, 0.5058823529411764, 0.7490196078431373),
 (0.6, 0.6, 0.6)]
```

```python
colors_fill = [(0.9843, 0.7059, 0.6824),
 (0.702, 0.8039, 0.8902),
 (0.8, 0.9216, 0.7725),
 (0.8706, 0.7961, 0.8941),
 (0.9961, 0.851, 0.651),
 (1, 1, 0.8),
 (0.898, 0.8471, 0.7412),
 (0.9922, 0.8549, 0.9255),
 (0.949, 0.949, 0.949)]
```

```python
data_C1C3 = []
with open('../data/likelihood_rotBIII_C1C3.dat', 'rt') as f:
    for l in f.readlines():
        data_C1C3.append([float(x) for x in l.split()])
data_C1C3 = np.array(data_C1C3)

dataRD_C1C3 = []
with open('../data/likelihood_rotBIII_C1C3_RD.dat', 'rt') as f:
    for l in f.readlines():
        dataRD_C1C3.append([float(x) for x in l.split()])
dataRD_C1C3 = np.array(dataRD_C1C3)

dataBKnunu_C1C3 = []
with open('../data/likelihood_rotBIII_C1C3_BKnunu.dat', 'rt') as f:
    for l in f.readlines():
        dataBKnunu_C1C3.append([float(x) for x in l.split()])
dataBKnunu_C1C3 = np.array(dataBKnunu_C1C3)
```

```python
c1range = np.linspace(-0.3, 0, 50)
c3range = np.linspace(-0.3, 0, 50)
c1_c1c3, c3_c1c3 = np.meshgrid(c1range, c3range)
```

```python
chi2 = -2*(data_C1C3-np.max(data_C1C3))
chi2BKnunu_c1c3 = -2*(dataBKnunu_C1C3-np.max(dataBKnunu_C1C3))
chi2RD_c1c3 = -2*(dataRD_C1C3-np.max(dataRD_C1C3))
```

```python
levels = [delta_chi2(n, 2) for n in (1,2)]
N = len(levels)
```

```python
plt.figure(figsize=(6,6))

plt.contour(scipy.ndimage.zoom(c1_c1c3, zoom=5, order=1), scipy.ndimage.zoom(c3_c1c3, zoom=5, order=1), scipy.ndimage.zoom(chi2RD_c1c3, zoom=5, order=2), levels=levels, colors=[colors_line[1]]*3, linewidths=0.6)

plt.contourf(scipy.ndimage.zoom(c1_c1c3, zoom=5, order=1), scipy.ndimage.zoom(c3_c1c3, zoom=5, order=1), scipy.ndimage.zoom(chi2RD_c1c3, zoom=5, order=2), levels=[-2,]+levels, colors=[colors_fill[1]+(max(1-n/(N+1), 0),) for n in range(1, N+1)])

plt.contour(scipy.ndimage.zoom(c1_c1c3, zoom=5, order=1), scipy.ndimage.zoom(c3_c1c3, zoom=5, order=1), scipy.ndimage.zoom(chi2BKnunu_c1c3, zoom=5, order=2), levels=levels, colors=[colors_line[3]]*3, linewidths=0.6)

plt.contourf(scipy.ndimage.zoom(c1_c1c3, zoom=5, order=1), scipy.ndimage.zoom(c3_c1c3, zoom=5, order=1), scipy.ndimage.zoom(chi2BKnunu_c1c3, zoom=5, order=2), levels=[-2,]+levels, colors=[colors_fill[3]+(max(1-n/(N+1), 0),) for n in range(1, N+1)])

plt.contour(scipy.ndimage.zoom(c1_c1c3, zoom=5, order=1), scipy.ndimage.zoom(c3_c1c3, zoom=5, order=1), scipy.ndimage.zoom(chi2, zoom=5, order=2), levels=levels, colors=[colors_line[4]]*3)

plt.contourf(scipy.ndimage.zoom(c1_c1c3, zoom=5, order=1), scipy.ndimage.zoom(c3_c1c3, zoom=5, order=1), scipy.ndimage.zoom(chi2, zoom=5, order=2), levels=[-2,]+levels, colors=[colors_fill[4]+(max(1-n/(N+1), 0),) for n in range(1, N+1)])

plt.annotate(r'$b\to c\ell\nu$', (-0.1, -0.173), color=colors_line[1], fontsize=15)
plt.annotate(r'$b\to s\nu\bar{\nu}$', (-0.2, -0.24), color=colors_line[3], fontsize=15)
plt.annotate('Global', (-0.09, -0.14), color=colors_line[4], fontsize=15)

plt.xlabel(r'$C_1$', fontsize=18)
plt.ylabel(r'$C_3$', fontsize=18)

plt.tight_layout(pad=0.5)
plt.savefig('likelihood_C1C3.pdf')
```
